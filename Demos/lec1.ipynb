```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
pollev_data = pd.read_csv('reason-for-taking-data110_summary_1467040.csv', 
                          skiprows=8, 
                          usecols=["Response options", "Count", "Percent"]) 
```

```python
pollev_data
```

```python
plt.xticks(rotation=-70)
sns.barplot(x=pollev_data['Response options'],y=pollev_data['Count'])
```

```python
ax=sns.barplot(orient='h',y=pollev_data['Response options'],
               x=pollev_data['Count']/pollev_data['Count'].sum()*100)
ax.set_xlabel('Percent of DATA110 students (%)')
plt.show()
```

```python
counts=pollev_data['Count']

plt.pie([counts[2],counts[0]+counts[1]+counts[3]+counts[4]],
        labels=['Students that care\nabout learning\nData Science','Students that do not care\nabout learning\nData Science'], 
        autopct='%.0f%%', 
        textprops={'fontsize': 14})
plt.show()
```

```python
plt.pie([counts[4],counts[0]+counts[1]+counts[2]+counts[3]],
        labels=['Students that care\nmore about cats than\nlearning Data Science','Students that do not\nlike animals'], 
        autopct='%.0f%%',
        textprops={'fontsize': 18})
plt.show()
```
